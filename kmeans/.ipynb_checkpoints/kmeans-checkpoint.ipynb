```python
import json
import pandas as pd
import numpy as np

with open('planes_data.json','r') as read_file:
    planes = json.load(read_file)
    read_file.close()

pd.read_json('planes_data.json')
```

```python
import matplotlib.pyplot as plt
plt.scatter([planes[i]['km'] for i in planes.keys()],[planes[i]['seats'] for i in planes.keys()])
plt.show()
```
